```python
# visualization of total cases through line chart
df=pd.read_csv("F:\CovidData.csv").head(10)
name=df['Name']
total_case=df['Cases - cumulative total']
# print(name)
# total_case
plt.figure(figsize=(15,5))
plt.plot(name,total_case,marker='s',c='r',markerfacecolor='black')
plt.ylabel('Number of cases in crore')
plt.xlabel('Country Name')
plt.yticks(range(25788387,765222932,99999999))
# # print(fig)
plt.show()
```
